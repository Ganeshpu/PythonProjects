# Matplotlib
In the below excercise I have tried ploting graphs using matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.arange(0,100)
y = x*2
z = x**2
```

* ** Plotted (x,y) on that axes and set the labels and titles**

```python
fig= plt.figure()
ax= fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('title')
```

** Created a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.**

```python
fig1= plt.figure()
ax1= fig1.add_axes([0,0,1,1])
ax2= fig1.add_axes([0.2,0.5,.2,.2])
```

**plot (x,y) on both axes**

```python
ax1.plot(x,y)
ax2.plot(x,y)

fig1
```

** Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
fig2=plt.figure()
ax1=fig2.add_axes([0,0,1,1])
ax2= fig2.add_axes([0.2,0.5,.4,.4])
ax1.plot(x,z)
ax1.set_xlabel('X')
ax1.set_ylabel('Y')

ax2.plot(x,y)
ax2.set_title('Zoom')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)
```

**Subplots**

```python
fig,axes= plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,color='blue',ls='--')
axes[1].plot(x,z,color='red',ls='-')
fig.tight_layout
```

** resize the plot by adding the figsize() argument in plt.subplots()**

```python
fig,axes= plt.subplots(nrows=1,ncols=2,figsize=(12,2))
axes[0].plot(x,y,color='blue',ls='--')
axes[1].plot(x,z,color='red',ls='-')
fig.tight_layout
```
