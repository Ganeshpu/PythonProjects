# Seaborn

## The Data

Working with a famous titanic data set for these graphs.

```python
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sb.set_style('whitegrid')
```

```python
titanic = sb.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sb.jointplot('fare','age',data=titanic)
```

```python
sb.distplot(titanic['fare'],bins=40,kde=False)
```

```python
sb.boxplot(x='class',y='age',data=titanic)
```

```python
sb.swarmplot(x='class',y='age',data=titanic)
```

```python
sb.countplot(x='sex',data=titanic)
```

```python
sb.heatmap(titanic.corr(),cmap='coolwarm')
```

```python
g = sb.FacetGrid(titanic, col='sex')
g.map(plt.hist, 'age', bins=30)
```
